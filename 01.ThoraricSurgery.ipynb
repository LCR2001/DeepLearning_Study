# 딥러닝 예제_1

< 폐암 환자 수술 후 생존 여부 예측 >

```python
# 딥러닝을 구동하는 데 필요한 케라스 함수 호출
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# 필요한 라이브러리 불러오기
import numpy as np
import tensorflow as tf
```

```python
# 실행할 때마다 같은 결과를 출력하기 위해 설정
np.random.seed(3)
tf.random.set_seed(3)
```

```python
# 준비된 수술 환자 데이터 불러오기
Data_set = np.loadtxt("ThoraricSurgery.csv", delimiter = ',')
```

```python
# 환자의 기록과 수술 결과 구분
X = Data_set[:, 0:17]
Y = Data_set[:, 17]
```

```python
# 딥러닝 구조를 결정(모델 설정 및 실행)
model = Sequential()
model.add(Dense(30, input_dim = 17, activation = 'relu'))  # 즉, 은닉층
model.add(Dense(1, activation = 'sigmoid'))   # 즉, 출력층
```

```python
# 딥러닝 실행1(오차함수 : 평균 제곱 오차)
model.compile(loss = 'mean_squared_error', optimizer = 'adam', metrics = ['accuracy'])
model.fit(X, Y, epochs = 100, batch_size = 10)
```

```python
# 딥러닝 실행2(오차함수 : 이항 교차 엔트로피 _ 클래스가 1 또는 0 두개이므로 사용 가능)
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
model.fit(X, Y, epochs = 100, batch_size = 10)
```

이외에도 다양한 오차함수
1. 평균 제곱 계열
    - 평균 제곱 오차
    - 평균 절대 오차(실제 값과 예측 값 차이의 절댓값 평균)
    - 평균 절대 백분율 오차(절댓값 오차를 절댓값으로 나눈 후 평균)
    - 평균 제곱 로그 오차(실제 값과 예측 값에 로그를 적용한 값의 차이를 제곱한 값의 평균)
2. 교차 엔트로피 계열
    - 범주형 교차 엔트로피(일반적인 분류)
    - 이항 교차 엔트로피(두개의 클래스 중에서 예측할 때)
